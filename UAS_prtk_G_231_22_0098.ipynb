```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from google.colab import files
uploaded = files.upload()
```

```python
df = pd.read_csv("stocks.csv")

df.head(20)
```

```python
# a
plt.figure(figsize=(27, 5))

plt.subplot(1, 3, 1)
plt.hist(df['Purchase Price($)'], bins=10, edgecolor='k')
plt.xlabel('Purchase Price')
plt.ylabel('Frequency')
plt.title('Histogram of Purchase Price')

plt.subplot(1, 3, 2)
plt.hist(df['Current Price($)'], bins=10, edgecolor='k')
plt.xlabel('Current Price')
plt.ylabel('Frequency')
plt.title('Histogram of Current Price')

plt.subplot(1, 3, 3)
plt.hist(df['Gain/Loss($)'], bins=10, edgecolor='k')
plt.xlabel('Gain/Loss')
plt.ylabel('Frequency')
plt.title('Histogram of Gain/Loss')

plt.tight_layout()
plt.show()
```

```python
#b
top_dividend_stocks = df.nlargest(5, 'Dividend/share($ )')[['Stock Name', 'Symbol']]
print("Top 5 stocks with highest Dividen / Share:")
print(top_dividend_stocks)
```

```python
#c
top_market_value_stocks = df.nlargest(5, 'Market Value($)')[['Stock Name', 'Symbol']]
print("Top 5 stocks with highest Market Value:")
print(top_market_value_stocks)
```

```python
#d
top_annual_yield_stocks = df.nlargest(7, 'Annual Yield(%)')[['Stock Name', 'Symbol']]
print("Top 7 stocks with highest Annual Yield:")
print(top_annual_yield_stocks)
```
